```python
#We will find the weak area where we can work to make more profit
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
sns.lineplot(x="Category", y="Quantity", data=a,ax=axes[0])
sns.lineplot(x="Category", y="Discount", data=a,ax=axes[1])
```

```python
#We will check the amount of profits and sales coming from items of different category
cw=a.groupby("Category")["Profit","Sales"].agg("sum")
print (cw)
cw.plot.bar()
plt.legend(loc="upper left")
plt.title("Category wise profit and sales")
```

```python
#We are getting more sales and profit from technology with less quantity of items
#we are getting less profit from furniture but quantity of item being sold is more than technology. So we need to decrease the discounts on furniture items 
```

**some other business problems**


```python
# Total profit and sales coming from each mode
rw=a.groupby("Ship Mode")["Profit", "Sales"].agg("sum")
print(rw)
rw.plot.bar()
plt.legend(loc="upper left")
plt.title("Ship Mode wise Profit and Sales")
```

```python
# checking quantity and discount on items from dofferent mode
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Quantity and Discounts on different mode")
sns.lineplot(x="Ship Mode",y="Discount",data=a,ax=axes[0])
sns.lineplot(x="Ship Mode",y="Quantity",data=a,ax=axes[1])
```

There are more sales on standard class but profit is not much with respect to sales. So, here we can decrease discounts on items being sold through standard class.So we can earn more profit from it.


On same we can increase discounts on the items being sold on the same day so that we could get more sales on same day. 

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Quantity and Discounts on items from different segments",size=15)
sns.lineplot(x="Segment",y="Discount",data=a,ax=axes[0])
sns.lineplot(x="Segment",y="Quantity",data=a,ax=axes[1])
plt.show()
```

Similarly, we need to increase discount on the items from Home Office Segment soo that we can get more sales from this segment

```python
fig,axes=plt.subplots(4,1,figsize=(14,6))
fig.suptitle("Quantity and Discount on different mode")
sns.barplot(x="Region",y="Quantity",data=a,hue="Category",ax=axes[0])
sns.barplot(x="Region",y="Discount",data=a,hue="Category",ax=axes[1])
sns.barplot(x="Region",y="Profit",data=a,hue="Category",ax=axes[2])
sns.barplot(x="Region",y="Sales",data=a,hue="Category",ax=axes[3])
```

Here we can see


Quantity of items are same in each region


Discount on items in central region is high


Profit is less in central region
